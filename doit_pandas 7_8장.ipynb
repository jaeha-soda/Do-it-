- 잘못 입력한 데이터 처리하기

```python
tips_sub_miss = tips.head(10)
tips_sub_miss.loc[[1,3,5,7], 'total_bill'] = 'missing'
print(tips_sub_miss)
```

```python
print(tips_sub_miss.dtypes)
```

```python
# tips_sub_miss['total_bill'].astype(float) # 당연히 오류가 뜬다.
```

```python
pd.to_numeric(tips_sub_miss['total_bill'])
```

ignore은 말 그대로 오류를 무시하는 거

```python
tips_sub_miss['total_bill']= pd.to_numeric(tips_sub_miss['total_bill'], errors='ignore')
print(tips_sub_miss)
```

```python
print(tips_sub_miss.dtypes)
```

erros 인자에 설정할 수 있는 값
- raise : 숫자로 변환할 수 없는 값이 있으면 오류 발생
- coerce : 숫자로 변환할 수 없는 값을 누락값으로 지정
- ignore : 아무 작업도 하지 않음

(erros 인자의 디폴트는 raise)

```python
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors='coerce')
print(tips_sub_miss)
```

```python
print(tips_sub_miss.dtypes)
```

def to_numeric(arg, errors='raise', downcast=None)

dowoncast는 정수, 실수와 같은 자료형을 더 작은 형태로 만들때 사용한다. 이를 다운캐스트라고 한다. downcast인자에는 interger, signed, unsigned, float 등의 값을 사용할 수 있다.

```python
tips_sub_miss['total_bill'] = pd.to_numeric(tips_sub_miss['total_bill'], errors='coerce', downcast='float')
print(tips_sub_miss.dtypes)
```

float64 -> float->32 로 바뀜. 메모리 공간이 2배 줄어들음. (빅데이터를 다룰때 유용함)
